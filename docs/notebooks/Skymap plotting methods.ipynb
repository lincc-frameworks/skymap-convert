# Plotting Methods

```python
from pathlib import Path

import skymap_convert
```

```python
package_root = Path.home() / "code" / "skymap-convert"

converted_skymap_path = package_root / "converted_skymaps" / "lsst_skymap"
converted_skymap_path

# TODO - I kind of want to just bundle the lsst skymap in, so users don't have to reason about a path.
```

```python
reader = skymap_convert.ConvertedSkymapReader(converted_skymap_path)
reader.summarize(allow_malformed=True)
```

```python
reader.help()
```

```python
reader.get_patch_vertices(3, 42)
```

```python
tract_id = 800

reader.plot_patches(
    [
        (tract_id, 0),
        (tract_id, 1),
        (tract_id, 9),
        (tract_id, 42),
        (tract_id, 90),
        (tract_id, 99),
    ],
    margin=0.05,
    tract_outer_boundaries=tract_id,
)
```

```python
reader.plot_patches([(3, 0), (3, 9), (3, 90), (3, 99)], margin=0.05, tract_outer_boundaries=3)
```
