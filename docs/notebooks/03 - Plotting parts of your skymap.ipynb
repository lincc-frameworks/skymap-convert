# Plotting Methods

```python
from pathlib import Path

import skymap_convert
```

```python
reader = skymap_convert.ConvertedSkymapReader(preset="lsst_skymap")

reader.summarize()
```

```python
reader.get_tract_vertices(42)
```

```python
reader.get_patch_vertices(1234, 56)
```

```python
reader.get_patch_vertices(3, 42)
```

```python
tract_id = 800

reader.plot_patches(
    [
        (tract_id, 0),
        (tract_id, 1),
        (tract_id, 9),
        (tract_id, 42),
        (tract_id, 90),
        (tract_id, 99),
    ],
    margin=0.05,
    tract_outer_boundaries=tract_id,
)
```

```python
reader.plot_patches([(3, 0), (3, 9), (3, 90), (3, 99)], margin=0.05, tract_outer_boundaries=3)
```

```python
reader.plot_patches(
    [
        (67, 90),
        (68, 99),
        (100, 2),
        (101, 9),
    ],
    margin=0.05,
    tract_outer_boundaries=[67, 68, 100, 101],
)
```
